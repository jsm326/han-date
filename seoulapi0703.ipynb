```python
import pandas as pd
import requests
```

```python
url = "http://openapi.seoul.go.kr:8088/4a496d74426a736d34327a6b6a4c55/json/InfoBizcenterApply/1/100/"
```

```python
r = requests.get(url)
```

```python
r.status_code
```

```python
data = r.json()

data
```

```python
jisan = data['InfoBizcenterApply']['row']
jisan
```

```python
jisan_df = pd.DataFrame(jisan)
jisan_df
def convert_date(date_str):
    try:
        # 'YYYY.MM.DD' 형식의 날짜 변환
        return pd.to_datetime(date_str, format='%Y.%m.%d').strftime('%Y.%m.%d')
    except ValueError:
        try:
            # 'YY.MM.DD' 형식의 날짜 변환
            return pd.to_datetime(date_str, format='%y.%m.%d').strftime('%Y.%m.%d')
        except ValueError:
            # 다른 형식의 날짜 변환 (일단 무시)
            return pd.NaT

jisan_df['APPROVAL_DATE'] = jisan_df['APPROVAL_DATE'].apply(convert_date)
jisan_df['COMPLE_DATE'] = jisan_df['COMPLE_DATE'].apply(convert_date)
```
